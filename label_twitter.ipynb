```python
import os
input_filename  = os.path.join("Data", "twitter", "python_tweets.json")
labels_filename = os.path.join("Data", "twitter", "python_classes.json")
```

```python
import json
tweets = []
with open(input_filename) as inf:
    for line in inf:
        if len(line.strip()) == 0:
            continue
        tweets.append(json.loads(line))
```

```python
labels = []
if os.path.exists(labels_filename):
    with open(labels_filename) as inf:
        labels = json.load(inf)
def get_next_tweet():
    return tweet_sample[len(labels)]['text']
tweet_sample = tweets
```

```python
%%javascript
function set_label(label){
    var kernel = IPython.notebook.kernel;
    kernel.execute("labels.append(" + label + ")");
    load_next_tweet();
}
function load_next_tweet(){
    var code_input = "get_next_tweet()";
    var kernel = IPython.notebook.kernel;
    var callbacks = {'iopub': {'output': handle_output} };
    kernel.execute(code_input, callbacks, {silent:false});
}
function handle_output(out){
    var res = out.content.data["text/plain"];
    $("div#tweet_text").html(res);
}
```

```python
%%html 
<div name="tweetbox">
    Instructions: Click in textbox. Enter a 1 if the tweet is relevant, enter 0 otherwise.<br>
        Tweet: <div id="tweet_text" value="text"></div><br>
               <input id="capture" type="text"></input><br>
        </div>
        <script>
        console.log("xxxxx");
        $("#capture").keypress(function(e) {
            console.log(e);
            if(e.which == 48) {
                console.log("48");
                set_label(0);
                $("#capture").val("");
            }else if (e.which == 49){
                console.log("49");
                set_label(1);
                $("#capture").val("");
            }
        });
        load_next_tweet();
        </script>
```

```python
def mark_text(text):
    print("Tweet: ", text)
    response = input("0/1: ")
    labels.append(int(response))

for tweet in tweet_sample:
    mark_text(tweet['text'])
    
```
